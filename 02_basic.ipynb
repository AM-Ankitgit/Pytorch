### 4 matrix operation

```python
f = torch.randint(size=(2,3),low=0,high=100)
g = torch.randint(size=(3,2),low=0,high=100)

print(f)
print(g)
```

```python
# matrix multiplication
torch.matmul(f,g)
```

```python
torch.dot(f,g)
```

```python
vector1 = torch.tensor([1, 2])
vector2 = torch.tensor([3, 4])

# dot product
torch.dot(vector1, vector2
          )
```
