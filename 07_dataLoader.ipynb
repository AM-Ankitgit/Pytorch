`To create batch of data  datalodaer class is used and for batch creation required the row by row data so for that dataset class is used`

```python
from sklearn.datasets import make_classification
```

```python
import torch
```

```python
X,y = make_classification(n_samples=10,n_features=2,n_informative=2,n_redundant=0,n_classes=2,random_state=42)
```

```python
X.shape
```

```python
y.shape
```

```python
X= torch.tensor(X,dtype=torch.float)
y= torch.tensor(y,dtype=torch.long)
```

```python
X
```

```python
y
```

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class CustomDataset(Dataset):

  def __init__(self, features, labels):

    self.features = features
    self.labels = labels

  def __len__(self):

    return self.features.shape[0]

  def __getitem__(self, index):

    return self.features[index], self.labels[index]
```

```python
dataset = CustomDataset(X,y)
```

```python
len(dataset)
```

```python
from torch.utils.data import DataLoader,Dataset
```

```python

dataloader = DataLoader(dataset=dataset,batch_size=2,shuffle=False)
```

```python

for batch_features,batch_labels in dataloader:
    print(batch_features)
    print(batch_labels)
```
